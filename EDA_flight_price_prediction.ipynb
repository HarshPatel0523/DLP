```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
type(df['Date_of_Journey'].str.split('/').str[0])
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split( ).str[0]
```

```python
df.head()
```

```python
df['Arrival Hours'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival Minutes'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Arrival Hours'] = df['Arrival Hours'].astype(int)
df['Arrival Minutes'] = df['Arrival Minutes'].astype(int)
```

```python
df.info()
```

```python
df['Dept_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dept_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dept_hour'] = df['Dept_hour'].astype(int)
df['Dept_min'] = df['Dept_min'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration Hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration Minutes'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.info()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```
